Problem: Count Even Numbers in a List

Problem statement:
Given a list of integers, return the count of even numbers.

nums = [1, 2, 3, 4, 6]
Output: 3

```python
def count_even(ls):
	count=0
	for i in ls:
		if i%2==0:
			count+=1
	return count
print(count_even([1,2,3,4,5,6]))
```

```python
def count_odd(ls):
	count=0
	for i in ls:
		if i%2!=0:
			count+=1
	return count
print(count_odd([1,2,3,4,5,6,7]))
```
