Task
Given a list of numbers, count:
how many are positive
how many are negative
how many are zero

```python
def counting(ls):
	count1=0
	count2=0
	count3=0
	for i in ls:
		if i==0:
			count1+=1
		elif i>0:
			count2+=1
		elif i<0:
			count3+=1
	print("Number of zeros:",count1)
	print("Number of positive numbers:",count2)
	print("Number of negative numbers:",count3)
counting([1,0,-1,3,0,-1,1,3,0,5,1,-1,4,0,-2,-4,5,2,-4,0])
```
