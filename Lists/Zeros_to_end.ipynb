```python
def zeros_toend(ls):
	for i in ls:
		if ls[i]==0:
			ls.append(ls.pop(i))
	return ls		
print(zeros_toend([0,1,2,0,3,0,4,5]))
```

```python
def move_zeros(nums):
    pos = 0  # position for non-zero

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[pos], nums[i] = nums[i], nums[pos]
            pos += 1

    return nums

print(move_zeros([0,1,2,0,3,0,4,5]))
```
