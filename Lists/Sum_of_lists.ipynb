Problem 1: Find the Sum of a List



Problem statement:
Given a list of integers, return the sum of all elements.

Example:
nums = [1, 2, 3, 4]
Output: 10

```python
ls=[1,2,3,4]
flag=0
for i in ls:
    flag=flag+i
print(flag)
```

wrap it in a function

```python
def sum_of_list(ls):
    flag=0
    for i in ls:
        flag = flag +i
    print(flag)
sum_of_list([1,2,3,4,5])
				
```
